```python
from collections import defaultdict
import csv

with open('sorted_zipmha22.txt', newline='') as csvfile:
    zip_locs = csv.reader(csvfile, delimiter=' ')

    zips = defaultdict()
    for zip, loc in zip_locs:
        zips[loc] = zip
```

```python
df['Zip'] = ''
df.head()
```

```python
for index, loc in df['Location'].iteritems():
    df['Zip'][index] = zips[loc]
df.head()
```

```python
with open('mhanames22.txt', newline='') as csvfile:
    cities = csv.reader(csvfile, delimiter=';')

    loc_cities = defaultdict()
    for loc, city in cities:
        cleaned_city = city.strip().split()
        cleaned_city = ' '.join(cleaned_city[:-1]) + ', ' + cleaned_city[-1]
        loc_cities[loc] = cleaned_city
```

```python
df['City'] = ''
for index, loc in df['Location'].iteritems():
    df['City'][index] = loc_cities[loc]
df.head()
```
